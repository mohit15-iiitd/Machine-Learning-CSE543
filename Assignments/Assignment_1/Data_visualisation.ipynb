```python
# Data Preparation
# Use StandardScaler to normalize df_x_train and df_x_test

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(df_x_train.drop(columns = ['label']))
x_test_scaled = scaler.fit_transform(df_x_test.drop(columns = ['label']))

print(f"Scaled Training Data without Label: {x_train_scaled}, Shape: {x_train_scaled.shape}")
print(f"Scaled Testing Data without Label: {x_test_scaled}, Shape: {x_test_scaled.shape}")
```

# Time Series Data

Household Power Consumption Prediction

(6 code/ comment blocks to be added)

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load the data
df_power = pd.read_csv("dataset4.csv", encoding = "ISO-8859-1", parse_dates = ['Datetime'], infer_datetime_format = True, index_col = "Datetime")
df_power.head()
```

```python
# Add code here
#Sort the dataframe according to date and time
df_power.sort_values(by=['Datetime'])
df_power.head()
```

```python
# Add code here
# Resample the hourly data to get data on a per week basis into 'df_weekly' dataframe and display first five rows of the 'df_weekly' dataframe
df_weekly = df_power.resample('W').sum()
df_weekly.head()
```

```python
# Add code here
# Sort the dataframe 'df_weekly' according to date and time
df_weekly = df_weekly.sort_values(by=['Datetime'])
df_weekly.head()
```

```python
#check the various trends or Decomposition of the time series
df_check_trend = df_weekly
decompose_data = seasonal_decompose(df_check_trend, model="additive")
decompose_data.plot()
plt.show()
```

```python
# Add comment below
# Comment on the above plot
```

This is seasonal_decompose, as the word suggest, it decomposes the time series data into three parts trend, seasonality, and residual. 

* **trend** shows the direction of the time series data over the period of time.
* **seasonality** shows a unique, and repeating pattern observed in intervals due to numerous factors i.e. monthly, weekly, etc.
* **residual** shows the irregular component consisting of fluctuations in time series after removing the previous components.

```python
df_weekly[:200].plot()
```

```python
# Add code here
# Use plot_acf and plot_pacf to plot acf and pcf both with lags=60 

plot_acf(df_weekly.values.squeeze(), lags=60)
plot_pacf(df_weekly.values.squeeze(), lags=60)
plt.show()
```

```python
# Add comment below
# comment on the plots (Comment on cyclicity and autocorrelation patterns)
```

The acf plot explains how closely the series present value relates to its past values.

The pcaf plot explains the correlation between the residuals and the next lag value rather than present with lag correlations like ACF.
